### Faça um programa que leia um número indeterminado de valores, correspondentes a notas, encerrando a entrada de dados quando for informado um valor igual a -1 (que não deve ser armazenado). Após esta entrada de dados, faça:
- Mostre a quantidade de valores que foram lidos;
- Exiba todos os valores na ordem em que foram informados, um ao lado do outro;
- Exiba todos os valores na ordem inversa à que foram informados, um abaixo do outro;
- Calcule e mostre a soma dos valores;
- Calcule e mostre a média dos valores;
- Calcule e mostre a quantidade de valores acima da média calculada;
- Calcule e mostre a quantidade de valores abaixo de sete;
- Encerre o programa com uma mensagem;

```python
list_of_notes = []
gibberish_list = []
counter = error_counter = 0
while True:
    counter += 1
    value = input(f"\n{'0' if counter < 10 else ''}{counter}-Type the note: ")
    try:
        if float(value) >= 0:
            list_of_notes.append(float(value))
    except ValueError:
        if value in '' or value.isspace():
            print("refused: fields can't be null!")
            gibberish_list.append(value)
        else:
            print(f"refused: {value} wasn't an expected float type!")
            gibberish_list.append(value)
        error_counter += 1
        counter -= 1
    else:
        if value == '-1':
            print(f'\nthe process has been finished.')
            print('\n------------------------------')
            break
        elif float(value) < 0:
            print(f"refused: negative numbers aren't supported!")
            gibberish_list.append(value)
            error_counter += 1
            counter -= 1
        else:
            print(f"succeeded validation!")
'example: 2.3  2.8  5.9  6.0  9.1  7.4  9.5  5.9  8.8  4.5'
print(f"\nRefused fields: {error_counter}")
print(f"\nRefused list: {gibberish_list}")
print(f"\nList of notes: {list_of_notes}")
```

```python
"v.1"
print(f'Mostre a quantidade de valores que foram lidos:\n{len(list_of_notes)}')

print('\nExiba todos os valores na ordem em que foram informados, um ao lado do outro:')
for values in list_of_notes:
    print(f'{values}', end='  ')

print('\n\nExiba todos os valores na ordem inversa à que foram informados, um abaixo do outro:')
for values in list_of_notes[::-1]:
    print(f'{values}')

print('\nCalcule e mostre a soma dos valores:')
print(sum(list_of_notes))

print('\nCalcule e mostre a média dos valores:')
media = sum(list_of_notes) / len(list_of_notes)
#print(round(media, 2))
print(media)

print('\nCalcule e mostre a quantidade de valores acima da média calculada:')
counter2 = 0
for values2 in list_of_notes:
    if values2 > media:
        print(f'{values2}', end='  ')
        counter2 += 1
print(f'\nQuantidade: {counter2}')

print('\nCalcule e mostre a quantidade de valores abaixo de sete:')
counter3 = 0
for values3 in list_of_notes:
    if values3 < 7:
        print(f'{values3}', end='  ')
        counter3 += 1
print(f'\nQuantidade: {counter3}')
```

```python
'v.2 - using comprehension list'

print(f'Mostre a quantidade de valores que foram lidos:\n{len(list_of_notes)}')

print('\nExiba todos os valores na ordem em que foram informados, um ao lado do outro:')
newlist1 = '  '.join([str(num) for num in list_of_notes])
print(newlist1)

print('\nExiba todos os valores na ordem inversa à que foram informados, um abaixo do outro:')
newlist2 = '\n'.join([str(num) for num in list_of_notes[::-1]])
print(f'{newlist2}')

print('\nCalcule e mostre a soma dos valores:')
print(sum(list_of_notes))

print('\nCalcule e mostre a média dos valores:')
media = (sum(list_of_notes) / len(list_of_notes))
print(round(media, 2))

print('\nCalcule e mostre a quantidade de valores acima da média calculada:')

newlist3 = '  '.join(list(str(num) for num in list_of_notes if num > media))
print(f'{newlist3}')
print(f'Quantidade: {len(list(num for num in list_of_notes if num > media))}')


print('\nCalcule e mostre a quantidade de valores abaixo de sete:')

newlist4 = '  '.join(list(str(num) for num in list_of_notes if num < 7))
print(f'{newlist4}')
print(f'Quantidade: {len(list(num for num in list_of_notes if num < 7))}')
```
