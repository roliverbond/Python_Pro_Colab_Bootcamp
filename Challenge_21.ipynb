### Faça um Programa para um caixa eletrônico. O programa deverá perguntar ao usuário a valor do saque e depois informar quantas notas de cada valor serão fornecidas. As notas disponíveis serão as de 1, 5, 10, 50 e 100 reais. O valor mínimo é de 10 reais e o máximo de 600 reais. O programa não deve se preocupar com a quantidade de notas existentes na máquina.
- Exemplo 1: Para sacar a quantia de 256 reais, o programa fornece duas notas de 100, uma nota de 50, uma nota de 5 e uma nota de 1;
- Exemplo 2: Para sacar a quantia de 399 reais, o programa fornece três notas de 100, uma nota de 50, quatro notas de 10, uma nota de 5 e quatro notas de 1.

```python
"V.1"
def cash_withdrawal(withdrawal: int):
    counter = 0
    a, b = divmod(withdrawal, 100)
    if a:
        counter += 1
    a_str = f"{a} nota{'s' if a > 1 else ''} de cem reais"
    b, c = divmod(b, 50)
    if b:
        counter += 1
    b_str = f"{b} nota{'s' if b > 1 else ''} de ciquenta reais"
    c, d = divmod(c, 10)
    if c:
        counter += 1
    c_str = f"{c} nota{'s' if c > 1 else ''} de dez reais"
    d, e = divmod(d, 5)
    if d:
        counter += 1
    d_str = f"{d} nota{'s' if d > 1 else ''} de cinco reais"
    if e:
        counter += 1
    e_str = f"{e} nota{'s' if e > 1 else ''} de um real"
    if counter == 0:
        print(f'O número "Zero" não é uma unidade monetária')
    else:
        print(f'O saque da quantia de R$ {withdrawal:.2f} será feita em:')
        if counter == 5:
            print(f'{a_str}, {b_str}, {c_str}, {d_str} e {e_str}.')
        elif counter == 4:
            if not e:
                print(f'{a_str}, {b_str}, {c_str} e {d_str}.')
            elif not d:
                print(f'{a_str}, {b_str}, {c_str} e {e_str}.')
            elif not c:
                print(f'{a_str}, {b_str}, {d_str} e {e_str}.')
            elif not b:
                print(f'{a_str}, {c_str}, {d_str} e {e_str}.')
            elif not a:
                print(f'{b_str}, {c_str}, {d_str} e {e_str}.')
        elif counter == 3:
            if not d and not e:
                print(f'{a_str}, {b_str} e {c_str}')
            elif not c and not e:
                print(f'{a_str}, {b_str} e {d_str}')
            elif not c and not d:
                print(f'{a_str}, {b_str} e {e_str}')
            elif not b and not e:
                print(f'{a_str}, {c_str} e {d_str}')
            elif not b and not d:
                print(f'{a_str}, {c_str} e {e_str}')
            elif not b and not c:
                print(f'{a_str}, {d_str} e {e_str}')
            elif not a and not e:
                print(f'{b_str}, {c_str} e {d_str}')
            elif not a and not d:
                print(f'{b_str}, {c_str} e {e_str}')
            elif not a and not c:
                print(f'{b_str}, {d_str} e {e_str}')
            elif not a and not b:
                print(f'{c_str}, {d_str} e {e_str}')
        elif counter == 2:
            if not c and not d and not e:
                print(f'{a_str} e {b_str}.')
            elif not b and not d and not e:
                print(f'{a_str} e {c_str}.')
            elif not b and not c and not e:
                print(f'{a_str} e {d_str}.')
            elif not b and not c and not d:
                print(f'{a_str} e {e_str}.')
                
            elif not a and not d and not e:
                print(f'{b_str} e {c_str}.')
            elif not a and not c and not e:
                print(f'{b_str} e {d_str}.')
            elif not a and not c and not d:
                print(f'{b_str} e {e_str}.')
                
            elif not a and not b and not e:
                print(f'{c_str} e {d_str}.')
            elif not a and not b and not d:
                print(f'{c_str} e {e_str}.')
            elif not a and not b and not c:
                print(f'{d_str} e {e_str}.')
        elif counter == 1:
            print(f"{a_str if a else ''}{b_str if b else ''}{c_str if c else ''}{d_str if d else ''}{e_str if e else ''}.")
    print('---------------------------------------------------------')


if __name__ == '__main__':
    testing_with = [256, 399, 326, 12, 300, 100, 320, 310, 305, 301, 101, 311, 111, 25, 20, 10, 21, 11, 1, 7, 16, 999, 1567, 567, 0]
    for values in testing_with:
        cash_withdrawal(values)
# print(f'a: {bool(a)}, b: {bool(b)}, c: {bool(c)}, d: {bool(d)}, e: {bool(e)}')
# print(f'counter: {counter}')
```

```python
"V.2"
def cash_withdrawal(withdrawal: int):
    a_str = b_str = c_str = d_str = e_str = ''
    counter = 0
    if withdrawal > 0:
        a, b = divmod(withdrawal, 100)
        if a:
            counter += 1
            if counter:
                a_str = f"{a} nota{'s' if a > 1 else ''} de cem reais"
        b, c = divmod(b, 50)
        if b:
            counter += 1
            if counter:
                b_str = f"{b} nota{'s' if b > 1 else ''} de ciquenta reais"
        c, d = divmod(c, 10)
        if c:
            counter += 1
            if counter:
                c_str = f"{c} nota{'s' if c > 1 else ''} de dez reais"
        d, e = divmod(d, 5)
        if d:
            counter += 1
            if counter:
                d_str = f"{d} nota{'s' if d > 1 else ''} de cinco reais"
        if e:
            counter += 1
            if counter:
                e_str = f"{e} nota{'s' if e > 1 else ''} de um real"
        
    if not withdrawal: # case "0"
        print(f'O número {withdrawal} (zero) não é uma unidade monetária.')
    elif withdrawal < 0: # case "negatives"
        print(f'O número {withdrawal} (negativo) não é suportado.')
    else:
        print(f'O saque da quantia de R$ {withdrawal:.2f} será feita em:'.replace('.', ','))

    if counter == 1:
        print(f'{a_str}{b_str}{c_str}{d_str}{e_str}.')

    elif counter == 2:
        two_parts = (f'{a_str} {b_str} {c_str} {d_str} {e_str}'.split())
        for c in range(len(two_parts)):                                # x.insert(5, 'e')
            print(f"{two_parts[c]}{' e' if c == 4 else ''}{'.' if c == len(two_parts)-1 else ''}", end=' ')

    elif counter == 3:
        three_parts = (f'{a_str} {b_str} {c_str} {d_str} {e_str}'.split())
        for c in range(len(three_parts)):
            print(f"{three_parts[c]}{',' if c == 4 else ''}{' e' if c == 9 else ''}{'.' if c == len(three_parts)-1 else ''}", end=' ')
    
    elif counter == 4:
        four_parts = (f'{a_str} {b_str} {c_str} {d_str} {e_str}'.split())
        for c in range(len(four_parts)):
            print(f"{four_parts[c]}{',' if c == 4 else ''}{',' if c == 9 else ''}{' e' if c == 14 else ''}{'.' if c == len(four_parts)-1 else ''}", end=' ')
    
    elif counter == 5:
        print(f'{a_str}, {b_str}, {c_str}, {d_str} e {e_str}.')
    print('\n---------------------------------------------------------')


if __name__ == '__main__':
    testing_with = [
            256, 399, 326, 12, 300, 100, 320, 310, 305, 301, 101, -34, 
            311, 111, 25, 20, 10, 21, 11, 1, 7, 16, 999, 1567, 567, 0,
            ]
    for values in testing_with:
        cash_withdrawal(values)
```

```python
"V.3 - final"
def cash_withdrawal(withdrawal: int):
    my_list = []
    if withdrawal > 0:
        a, b = divmod(withdrawal, 100)
        if a: my_list.append(f"{a} nota{'s' if a > 1 else ''} de cem reais")
        b, c = divmod(b, 50)
        if b: my_list.append(f"{b} nota{'s' if b > 1 else ''} de ciquenta reais")
        c, d = divmod(c, 10)
        if c: my_list.append(f"{c} nota{'s' if c > 1 else ''} de dez reais")
        d, e = divmod(d, 5)
        if d: my_list.append(f"{d} nota{'s' if d > 1 else ''} de cinco reais")
        if e: my_list.append(f"{e} nota{'s' if e > 1 else ''} de um real")
    
    if not withdrawal: # case "0"
        print(f'O número {withdrawal} (zero) não é uma unidade monetária.')
    elif withdrawal < 0: # case "negatives"
        print(f'O número {withdrawal} (negativo) não é suportado.')
    else:
        print(f'O saque da quantia de R$ {withdrawal:.2f} será feita em:'.replace('.', ','))

    for c in range(len(my_list)):
        print(f'{my_list[c]}', end='')
        if c == len(my_list) - 2:
            print(f'{" e "}', end='')
        elif c == len(my_list) - 1:
            print(f'{"."}')
        else:
            print(f'{", "}', end='')
    print('---------------------------------------------------------')


if __name__ == '__main__':
    testing_with = [
            256, 399, 326, 12, 300, 100, 320, 310, 305, 301, 101, -34, 
            311, 111, 25, 20, 10, 21, 11, 1, 7, 16, 999, 1567, 567, 0,
            ]
    for values in testing_with:
        cash_withdrawal(values)
```
