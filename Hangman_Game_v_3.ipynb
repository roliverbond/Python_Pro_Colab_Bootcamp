Jogo de Forca. Desenvolva um jogo da forca. O programa terá uma lista de palavras lidas de um arquivo texto e escolherá uma aleatoriamente. O jogador poderá errar 6 vezes antes de ser enforcado.
```
Digite uma letra: A
-> Você errou pela 1ª vez. Tente de novo!

Digite uma letra: O
A palavra é: _ _ _ _ O

Digite uma letra: E
A palavra é: _ E _ _ O

Digite uma letra: S
-> Você errou pela 2ª vez. Tente de novo!
```


````python
def drawing_hanged_man(errors):
    print("  _______     ")
    print(" |/      |    ")

    if(errors == 1):
        print(" |      (_)   ")
        print(" |            ")
        print(" |            ")
        print(" |            ")

    if(errors == 2):
        print(" |      (_)   ")
        print(" |      \     ")
        print(" |            ")
        print(" |            ")

    if(errors == 3):
        print(" |      (_)   ")
        print(" |      \|    ")
        print(" |            ")
        print(" |            ")

    if(errors == 4):
        print(" |      (_)   ")
        print(" |      \|/   ")
        print(" |            ")
        print(" |            ")

    if(errors == 5):
        print(" |      (_)   ")
        print(" |      \|/   ")
        print(" |       |    ")
        print(" |            ")

    if(errors == 6):
        print(" |      (_)   ")
        print(" |      \|/   ")
        print(" |       |    ")
        print(" |      /     ")

    if (errors == 7):
        print(" |      (_)   ")
        print(" |      \|/   ")
        print(" |       |    ")
        print(" |      / \   ")

    print(" |            ")
    print("_|___         ")
    print()


def winner_message():
    print("Congratulations, you win!")
    print("       ___________      ")
    print("      '._==_==_=_.'     ")
    print("      .-\\:      /-.    ")
    print("     | (|:.     |) |    ")
    print("      '-|:.     |-'     ")
    print("        \\::.    /      ")
    print("         '::. .'        ")
    print("           ) (          ")
    print("         _.' '._        ")
    print("        '-------'       ")


def loser_message(secret: string):
    print("You're a hanged man, you lose!")
    print(f"The secret word is {secret}")
    print("    _______________      ")
    print("   /               \     ")
    print("  /                 \    ")
    print("//                   \/\ ")
    print("\|   XXXX     XXXX   | / ")
    print(" |   XXXX     XXXX   |/  ")
    print(" |   XXX       XXX   |   ")
    print(" |                   |   ")
    print(" \__      XXX      __/   ")
    print("   |\     XXX     /|     ")
    print("   | |           | |     ")
    print("   | I I I I I I I |     ")
    print("   |  I I I I I I  |     ")
    print("   \_             _/     ")
    print("     \_         _/       ")
    print("       \_______/         ")
````

````python
from time import sleep
from random import shuffle, randint


def shuffle_words() -> str:
    fruit_list = [
        "apple", "banana", "cherry", 'raspberry', 'watermelon', 'Pineapple', 'Blackberry', 'fig','Coconut',
        'Cashew', 'Avocado', 'Plum', 'Apricot', 'Guava', 'lime', 'Kiwi fruit', 'Orange', 'Lemon', 'Soursop',
        "Mango", 'Passion fruit', 'Strawberry', 'Peach', 'Pear', 'Grape', 'Tangerine', 'Star fruit','Melon',
        ]
    shuffle(fruit_list)
    return fruit_list[randint(0, len(fruit_list))].upper()


attempted_chars = []
correct_words = []
secret_word = shuffle_words()
for chars in secret_word:
    correct_words.append('___' if chars.isalpha() else ' ')
#print(f"\n{''.join(correct_words)}\n")
for c in correct_words:
    print(f'{c}', end=' ')
    sleep(0.5)
print('\n')
c = 0
while c < 7:
    while True:
        string = input('letter: ').replace(" ", "").upper()
        

        if string.isdigit() or string == '':
            print('- invalid: Only letters!\n')
        
        elif string not in secret_word:
            print(f"- the letter '{string}' doesn't belong to the secret word. You have {abs(c-6)} more attempt(s).\n")
            attempted_chars.append(string)
            print(f"- unsuitable letters: {', '.join(attempted_chars)}")
            c += 1
            drawing_hanged_man(c)
            if c == 7:
                loser_message(secret_word)
                break
        else:
            break
        
        sleep(1.5)
    print('')

    index = 0
    for letter in secret_word:
        if string == letter:
            correct_words[index] = f' {letter} '
        index += 1   
    sleep(0.5)
    print(f"{' '.join(correct_words)}\n")
    sleep(1)
    if not '___' in correct_words:
        winner_message()
        break        
sleep(2)
print('\nEnd of the program.')
````
