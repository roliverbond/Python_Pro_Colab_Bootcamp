```python
from time import sleep
from random import shuffle, randint


def drawing_hanged_man(errors):
    print("  _______     ")
    print(" |/      |    ")

    if errors == 1:
        print(" |      (_)   ")
        print(" |            ")
        print(" |            ")
        print(" |            ")

    if errors == 2:
        print(" |      (_)   ")
        print(" |      \     ")
        print(" |            ")
        print(" |            ")

    if errors == 3:
        print(" |      (_)   ")
        print(" |      \|    ")
        print(" |            ")
        print(" |            ")

    if errors == 4:
        print(" |      (_)   ")
        print(" |      \|/   ")
        print(" |            ")
        print(" |            ")

    if errors == 5:
        print(" |      (_)   ")
        print(" |      \|/   ")
        print(" |       |    ")
        print(" |            ")

    if errors == 6:
        print(" |      (_)   ")
        print(" |      \|/   ")
        print(" |       |    ")
        print(" |      /     ")

    if errors == 7:
        print(" |      (_)   ")
        print(" |      \|/   ")
        print(" |       |    ")
        print(" |      / \   ")

    print(" |            ")
    print("_|___         ")
    print()


def winner_message():
    print("Congratulations, you win!")
    print("       ___________      ")
    print("      '._==_==_=_.'     ")
    print("      .-\\:      /-.    ")
    print("     | (|:.     |) |    ")
    print("      '-|:.     |-'     ")
    print("        \\::.    /      ")
    print("         '::. .'        ")
    print("           ) (          ")
    print("         _.' '._        ")
    print("        '-------'       ")


def loser_message(secret: str):
    print("You're a hanged man, you lose!")
    print(f"The secret word was {secret}")
    print("    _______________      ")
    print("   /               \     ")
    print("  /                 \    ")
    print("//                   \/\ ")
    print("\|   XXXX     XXXX   | / ")
    print(" |   XXXX     XXXX   |/  ")
    print(" |   XXX       XXX   |   ")
    print(" |                   |   ")
    print(" \__      XXX      __/   ")
    print("   |\     XXX     /|     ")
    print("   | |           | |     ")
    print("   | I I I I I I I |     ")
    print("   |  I I I I I I  |     ")
    print("   \_             _/     ")
    print("     \_         _/       ")
    print("       \_______/         ")

```

```python
def shuffle_words() -> str:
    fruit_list = [
        "apple", "banana", "cherry", 'raspberry', 'watermelon', 'Pineapple', 'Blackberry', 'fig', 'Coconut',
        'Cashew', 'Avocado', 'Plum', 'Apricot', 'Guava', 'lime', 'Kiwi fruit', 'Orange', 'Lemon', 'Soursop',
        "Mango", 'Passion fruit', 'Strawberry', 'Peach', 'Pear', 'Grape', 'Tangerine', 'Star fruit', 'Melon',
    ]
    shuffle(fruit_list)
    return fruit_list[randint(0, len(fruit_list))].upper()


attempted_chars = []
correct_words = []
secret_word = shuffle_words()
ctr_error = 0
string = ''

for chars in secret_word:
    correct_words.append('___' if chars.isalpha() else ' ')

print("\nWelcome to the Hangman's Game!\n")


while True:
    index = 0
    for letter in secret_word:
        if string == letter:
            correct_words[index] = f' {letter} '
        index += 1
    sleep(0.5)

    for c in correct_words:
        sleep(0.25)
        print(f'{c}', end=' ')
    sleep(0.25)
    print('\n')

    if '___' not in correct_words:
        sleep(1.5)
        winner_message()
        sleep(3)
        break

    string = input('- type a letter: ').replace(' ', '').upper()

    if string.isdigit() or string == '':
        print('- invalid: letters only!\n')

    elif string not in secret_word:
        sleep(0.5)
        print(f"- letter '{string}' doesn't belong to the secret word. You have {abs(ctr_error - 6)} more attempt(s).")
        attempted_chars.append(string)
        sleep(0.5)
        print(f"- unsuitable letters: {', '.join(attempted_chars)}")
        ctr_error += 1
        sleep(1)
        drawing_hanged_man(ctr_error)

        if ctr_error == 7:
            sleep(1.5)
            loser_message(secret_word)
            sleep(3)
            break

    print('')
sleep(3)
print('\nEnd of the program.')
```
